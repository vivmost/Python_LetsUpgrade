```python
from random import randint
option = ["rock" , "paper" , "scissor"]

computer = option[randint(0,2)]
computer_points = 0
player_points = 0
ties = 0

choice = True

while choice == True:
    player = input(" \n   To play choose from:- \n 1.rock   \n 2.paper    \n 3.scissor   \n  And to exit type  'EXIT'  ")
    if player == computer:
        print("it's a tie this time ")
        ties += 1
        
    elif player == "rock":
        if computer == "paper":
            print("you loose this time ")
            computer_points += 1
        elif computer == player:
            print("it's a tie ")
            ties += 1
        else:
            print("you win this time ")
            player_points += 1
            
    elif player == "paper":
        if computer == "scissor":
            print("you loose this time ")
            computer_points += 1
        elif computer == player:
            print("it's a tie ")
            ties += 1
        else:
            print("you win this time ")
            player_points += 1
    
    elif player == "scissor":
        if  computer == "rock":
            print("you loose this time ")
            computer_points += 1
        elif computer == player:
            print("it's a tie ")
            ties += 1
        else:
            print("you won this time ")
            player_points += 1
            
    elif player == "EXIT":
        break
            
    else:
        print("Invalid input, Try again ")
        
    
print("\n")
print("<---------------game board---------------->")   
print("the points of the player are : " , player_points)
print("the points of the computer are : " , computer_points)
print("the no, of times the game tied :" , ties)
print("\n")

    
if (player_points > computer_points):
    print("you won the complete game ")
elif(player_points == computer_points):
    print("nobody won the complete game , it's a tie")
else:
    print("better luck next time , you loose")
    
print("<----------------------------------------->") 
```
