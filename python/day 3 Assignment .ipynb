```python
lower = 1042000
upper = 702648265

for num in range(lower, upper + 1):
    order = len(str(num))
    result = 0
    temp = num
    while temp > 0:
        digit = temp % 10
        result += digit ** order
        temp //= 10
        
    if(num == result):
        print("the first armstrong no. is : " , num)
        break
    else:
        continue
    
```
